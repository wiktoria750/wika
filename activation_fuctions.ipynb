## Funkcje Aktywacji

### ReLu Function

$f(x) = max(x, 0)$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def max_relu(x):
  return max(x, 0)
```

```python
for i in [-10, -5, 0, 5, 10]:
  print(max_relu(i))
```

```python
data = np.arange(-5, 5, 0.1)
max_relu_data = [max_relu(x) for x in data]

plt.plot(data, max_relu_data)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## Sigmoid Function

### $f(x) = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))
```

```python
for i in [-5, -3, -1, 0, 1, 3, 5]:
  print(sigmoid(i))
```

```python
sigmoid_data = [sigmoid(x) for x in data]

plt.plot(data, sigmoid_data)
plt.title('Sigmoid Function')
```

## Softmax Function

### $S(y_{i}) = \frac{e^{y_{i}}}{\sum_{j=1}^{i} e^{y_{j}}}$

```python
data = np.random.randn(3, 5)
data
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
res = np.exp(df) / np.sum(np.exp(df), axis=0)
res
```

## Tan h

### $tanh(x) = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
for i in [-2, -1, 0, 1, 2]:
  print(tanh(i))
```

```python
data = np.arange(-5, 5, 0.1)
tanh_data = [tanh(x) for x in data]

plt.plot(data, tanh_data)
plt.title('Tanh Function')
```
